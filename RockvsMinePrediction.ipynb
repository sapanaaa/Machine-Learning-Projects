```python
#IMPORTING LIBRARIES 
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading the dataset to a pandas dataset
sonar_data = pd.read_csv('Downloads/Copy of sonar data.csv', header=None)
```

```python
#HEAD FUNCTION SHOWS FIRST 5 ROWS
sonar_data.head()

```

```python
# shape gives the number of rows and columns
sonar_data.shape
```

```python
#value_counts lists the number of variables are in the dataset
sonar_data[60].value_counts()
```

```python
sonar_data.groupby(60).mean()
```

```python
#separating data and labels
X=sonar_data.drop(columns=60, axis=1)
Y= sonar_data[60]
```

```python
print(X)
print(Y)
```

```python
#training and testing the data
#we are going to split into train and test data using train_test_split function
#test_size gives the percentage of data assigned for testing
#stratify splits the data on the basis of rocks and mines

Xtrain, Xtest, Ytrain, Ytest= train_test_split(X,Y, test_size= 0.1, stratify=Y, random_state=1)
```

```python
print(X.shape, Xtrain.shape, Xtest.shape)
```

```python
print(Xtrain)
print(Ytrain)
```

```python
#MODEL TRAINING USING LOGISTIC REGRESSION WITH TRAINING DATA
model= LogisticRegression()
```

```python
model.fit(Xtrain, Ytrain)
```

```python
#MODEL EVALUATION
#ACCURACY ON TRAINING DATA
Xtrain_prediction= model.predict(Xtrain)
training_data_accuracy= accuracy_score(Xtrain_prediction, Ytrain)
#here Xtrain is the predicted data and Ytrain is expected data
```

```python
print("accuracy on our training data is:", training_data_accuracy)
```

```python
#ACCURACY ON TEST DATA
Xtest_prediction= model.predict(Xtest)
test_data_accuracy= accuracy_score(Xtest_prediction, Ytest)
#here Xtest is the predicted data and Ytest is expected data
```

```python
print("accuracy on our test data is:", test_data_accuracy)
```
